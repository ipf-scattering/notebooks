```python
from ipychord3 import sf, utils
```

### without rotation

```python
p = utils.TestPattern().make_pattern(0)
```

```python
sf.show(p)
```

```python
harmonized = sf.harmony(p, p['beam_position'])
```

```python
sf.show(harmonized)
```

### with rotation

```python
angle = 13
p_rot = utils.TestPattern().make_pattern(angle)
```

```python
sf.show(p_rot)
```

```python
harmonized_rot = sf.harmony(p_rot, p_rot['beam_position'], angle=-angle)
```

```python
sf.show(harmonized_rot)
```
